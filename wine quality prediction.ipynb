```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
ss = StandardScaler()
X = pd.DataFrame(ss.fit_transform(X),columns=df.columns[:-1])
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
sort_ind = rf_model.feature_importances_.argsort()
plt.barh(df.columns[sort_ind],rf_model.feature_importances_[sort_ind])
```

```python
y_pred_train = rf_model.predict(x_train)
y_pred_test = rf_model.predict(x_test)
print('Accuracy score of Random Forest with training dataset is : ', accuracy_score(y_pred_train,y_train))
print('Accuracy score of Random Forest with testing dataset is : ', accuracy_score(y_pred_test,y_test))
```

```python
cross_val_score(rf_model,X,y,cv=10).mean()
```

```python
y.value_counts()
```

```python
quality_ = []
for i in df['quality']:
    if(i==3 or i==4 or i==5):
        quality_.append(0)
    else:
        quality_.append(1)
        
df['quality_'] = quality_
```

```python
df['quality_'].value_counts()
```

```python
X_ = X
y_ = df.iloc[:,-1]
```

```python
x_train1,x_test1,y_train1,y_test1 = train_test_split(X_,y_,test_size=0.25,random_state=0)
```

```python
rf_model1 = RandomForestClassifier()
rf_model1.fit(x_train1,y_train1)
```

```python
y_pred_train1 = rf_model1.predict(x_train1)
y_pred_test1 = rf_model1.predict(x_test1)
print('Accuracy score of Random Forest with training dataset is : ', accuracy_score(y_pred_train1,y_train1))
print('Accuracy score of Random Forest with testing dataset is : ', accuracy_score(y_pred_test1,y_test1))
```

```python
cross_val_score(rf_model1,X_,y_,cv=10).mean()
```

```python
print(confusion_matrix(y_pred_test1,y_test1))
```

```python
print(classification_report(y_pred_test1,y_test1))
```

```python
param_dict = {
    'n_estimators': [200,300,500,1000],
    'min_samples_leaf':[2,5,10],
    'min_samples_split': [8, 10, 12]
}
grid_cv = GridSearchCV(rf_model1,param_dict,cv=5,scoring='accuracy')
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.best_params_
```

```python
rf_model2 = RandomForestClassifier(n_estimators=300,min_samples_leaf=2,min_samples_split=8)
rf_model2.fit(x_train1,y_train1)
```

```python
y_pred_train2 = rf_model2.predict(x_train1)
y_pred_test2 = rf_model2.predict(x_test1)
print('Accuracy score of Random Forest with training dataset is : ', accuracy_score(y_pred_train2,y_train1))
print('Accuracy score of Random Forest with testing dataset is : ', accuracy_score(y_pred_test2,y_test1))
```

```python
cross_val_score(rf_model2,X_,y_,cv=10).mean()
```

```python
print(confusion_matrix(y_pred_test2,y_test1))
```

```python
print(classification_report(y_pred_test2,y_test1))
```

```python
y_true_train,y_true_test = y_train1,y_test1
y_scores_train1,y_scores_test1 = y_pred_train1,y_pred_test1
y_scores_train2,y_scores_test2 = y_pred_train2,y_pred_test2
```

```python
fpr_train1, tpr_train1, thresholds_train1 = roc_curve(y_true_train, y_scores_train1)
fpr_test1, tpr_test1, thresholds_test1 = roc_curve(y_true_test, y_scores_test1)

fpr_train2, tpr_train2, thresholds_train2 = roc_curve(y_true_train, y_scores_train2)
fpr_test2, tpr_test2, thresholds_test2 = roc_curve(y_true_test, y_scores_test2)
```

```python
auc_train1 = roc_auc_score(y_true_train, y_scores_train1)
auc_test1 = roc_auc_score(y_true_test, y_scores_test1)

auc_train2 = roc_auc_score(y_true_train, y_scores_train2)
auc_test2 = roc_auc_score(y_true_test, y_scores_test2)
```

```python
plt.plot(fpr_train1, tpr_train1, label='Training Data (AUC = {:.2f})'.format(auc_train1))
plt.plot(fpr_test1, tpr_test1, label='Testing Data (AUC = {:.2f})'.format(auc_test1))
plt.plot(fpr_train2, tpr_train2, label='Training Data (AUC = {:.2f})'.format(auc_train2))
plt.plot(fpr_test2, tpr_test2, label='Testing Data (AUC = {:.2f})'.format(auc_test2))
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
